# Employee Bonuses
### FirstChoice Bank · South Street Branch

At the bank, customers can rate tellers after an interaction, and tellers earn an annual bonus based on their average customer rating for the year.

This notebook holds the branch's five tellers, applies the bank's ranged bonus structure, and prints each teller's bonus.

*This is the notebook we put under version control in the Anatomy of a Repository video, and work on through the rest of the module.*

## The bonus structure

| Average rating | Bonus |
|---|---|
| 4.5 to 5 | 17% of salary |
| 4 up to 4.5 | 10% of salary |
| 3 up to 4 | 5% of salary |
| 1.5 up to 3 | 2% of salary |
| below 1.5 | 0% |

## List of tellers

The five tellers at the South Street Branch, with this year's average customer rating and their salary:

```python
employees = [
    {"name": "Alice",   "rating": 4.88, "salary": 67004.55},
    {"name": "Bob",     "rating": 4.201,  "salary": 52300.01},
    {"name": "Charlie", "rating": 3.26, "salary": 46325.65},
    {"name": "David",   "rating": 2.33, "salary": 45213.99},
    {"name": "Eve",     "rating": 1.06, "salary": 35122.37}
]
```

```python
#option to manually add additional teller

manual_input = input("Would you like to add a teller? Enter Y or N. ")

if manual_input.strip().upper() == 'Y':
  new_name = input("Employee first name: ")
  new_rating = float(input("Employee average customer rating: "))
  new_salary = float(input("Employee salary: "))
  employees.append({
        "name": new_name,
        "rating": new_rating,
        "salary": new_salary
    })

  print(f"{new_name} has been added with a customer rating of {new_rating} and a salary of {new_salary}. You may continue to run this notebook.")
else:
  print("Thank you. You may continue to rub this notebook.")
```

## Calculate each teller's bonus

```python
def bonus_calculator(emp_list):
  for employee in emp_list:
      t_bonus = 0
      name = employee["name"]
      t_rating = employee["rating"]
      t_salary = employee["salary"]
      if t_rating > 4.5:
          t_bonus = t_salary * (.17)
      elif t_rating >= 4:
          t_bonus = t_salary * (.10)
      elif t_rating >= 3:
          t_bonus = t_salary * (.05)
      elif t_rating >= 1.5:
          t_bonus = t_salary * (.02)
      else:
          t_bonus = 0
      print(f"{name} has earned a ${round(t_bonus,2)} bonus.");

bonus_calculator(employees)
```
